```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def getSentimentScores(body):
    sid_obj = SentimentIntensityAnalyzer()
  
    sentiment_dict = sid_obj.polarity_scores(body)
    
    negScore = sentiment_dict['neg']
    neuScore = sentiment_dict['neu']
    posScore = sentiment_dict['pos']
    
    return [negScore, neuScore, posScore]

print(getSentimentScores("I ae3h hw436w () ©ðŸ˜‹ bad bad bad"))
```

```python
def sentimentAnalysisAttachment(df):
    body = list(df['comment_body'])
    
    negativeScores = []
    neutralScores = []
    positiveScores = []
    
    for comment in body:
        scores = getSentimentScores(str(comment))
        negativeScores.append(scores[0])
        neutralScores.append(scores[1])
        positiveScores.append(scores[2])
            
    
    print(negativeScores)
    print(neutralScores)
    print(positiveScores)
    df["negative_Score"] = negativeScores
    df["neutral_Score"] = neutralScores
    df["positive_Score"] = positiveScores
    
    return df;
```
